### 2. Build a Financial Scoring Model:

```python
# Scoring Function
def calculate_score(row):
    # Factor Calculations
    savings_to_income = (row["Savings"] / row["Income"]) * 100
    expenses_to_income = (row["Monthly Expenses"] / row["Income"]) * 100
    loans_to_income = (row["Loan Payments"] / row["Income"]) * 100
    credit_card_trend_penalty = min(100, row["Credit Card Spending"] / row["Income"] * 10)
    discretionary_penalty = 10 if row["Amount"] > (row["Income"] * 0.3) else 0
    goals_met = row["Financial Goals Met (%)"]

    # Weighted Score
    score = (
        (savings_to_income * 0.3) +
        ((100 - expenses_to_income) * 0.25) +
        ((100 - loans_to_income) * 0.15) +
        ((100 - credit_card_trend_penalty) * 0.1) +
        ((100 - discretionary_penalty) * 0.1) +
        (goals_met * 0.1)
    )
    return min(max(score, 0), 100)  # Ensure score is between 0 and 100

# Apply Scoring
df["Financial Score"] = df.apply(calculate_score, axis=1)
df["Financial Score"]
```

#### Key Observations:

**Score Range:**

Scores are falling within the expected range of 0 to 100, which is good.
The values seem to vary across the dataset, which suggests the model is taking into account the different factors (income, expenses, loans, credit card usage, discretionary spending, and financial goals met).

**Score Interpretation:**

Higher scores likely indicate better financial health, while lower scores suggest areas that need improvement.

**The factors contributing to the score are:**

Savings-to-Income Ratio: This contributes 30% of the score. Higher savings relative to income increases the score.

Monthly Expenses as % of Income: Lower expenses relative to income improve the score (it’s subtracted from 100).

Loan Payments as % of Income: Similar to expenses, lower loan payments as a percentage of income will increase the score.

Credit Card Spending: More credit card spending relative to income results in a deduction from the score.

Discretionary Spending (Travel/Entertainment): Higher discretionary spending (if over 30% of income) leads to a penalty.

Financial Goals Met: The percentage of financial goals met directly adds to the score.

**Score Variability:**

Scores such as 68, 78, 61, 88 indicate some members are performing better financially than others.
The factors with the highest weight (e.g., Savings-to-Income and Monthly Expenses) are likely driving the major differences in the scores.

```python
def categorize_log_score(log_score):
    if log_score >= 4.5:
        return 'Excellent'
    elif log_score >= 4.0:
        return 'Good'
    elif log_score >= 3.7:
        return 'Average'
    else:
        return 'Poor'

df['Score Category'] = df['Log Financial Score'].apply(categorize_log_score)

# Check the updated categories
print(df[['Family ID', 'Member ID', 'Financial Score', 'Log Financial Score', 'Score Category']].head())
```

```python
def generate_recommendations(row):
    recommendations = []

    # Savings Recommendation
    savings_to_income = (row["Savings"] / row["Income"]) * 100
    if savings_to_income < 10:  # Less than 10% savings-to-income ratio is considered low
        recommendations.append("Increase your savings by saving at least 10% of your monthly income.")

    # Expenses Recommendation
    expenses_to_income = (row["Monthly Expenses"] / row["Income"]) * 100
    if expenses_to_income > 50:  # If expenses are greater than 50% of income, it's considered high
        recommendations.append("Try reducing your monthly expenses, particularly in discretionary categories like travel or entertainment.")

    # Loan Payments Recommendation
    loans_to_income = (row["Loan Payments"] / row["Income"]) * 100
    if loans_to_income > 20:  # If loan payments exceed 20% of income, consider consolidating or refinancing
        recommendations.append("Consider consolidating your loans or refinancing to reduce your monthly payments.")

    # Credit Card Spending Recommendation
    if row["Credit Card Spending"] > row["Income"] * 0.3:  # More than 30% of income spent on credit cards is high
        recommendations.append("Reduce your credit card spending and focus on paying off high-interest debt first.")

    # Financial Goals Recommendation
    if row["Financial Goals Met (%)"] < 70:  # If less than 70% of financial goals are met, there's room for improvement
        recommendations.append("Focus on achieving more of your financial goals by budgeting and setting clear priorities.")

    return recommendations

# Apply the recommendation function to the dataset
df['Recommendations'] = df.apply(generate_recommendations, axis=1)

# View the results with recommendations
df[['Family ID','Member ID', 'Financial Score', 'Recommendations']]
```

```python
# Aggregate duplicates by taking the sum (you can also use mean, etc.)
# Select only numeric columns for aggregation
numeric_columns = df.select_dtypes(include=["number"])

# Aggregate numeric columns by summing
df_aggregated = numeric_columns.groupby(df.index).sum()

# If you need to keep non-numeric columns, you can handle them separately (e.g., keep the first or use other aggregation methods)
non_numeric_columns = df.select_dtypes(exclude=["number"])
non_numeric_aggregated = non_numeric_columns.groupby(df.index).first()

# Combine the aggregated numeric and non-numeric data
df_final = pd.concat([df_aggregated, non_numeric_aggregated], axis=1)

print(df_final.head())
```

```python
from statsmodels.tsa.stattools import adfuller

adf_test = adfuller(df_final['Monthly Expenses'])
print(f"ADF Statistic: {adf_test[0]}, p-value: {adf_test[1]}")
if adf_test[1] > 0.05:
    print("Data is not stationary. Consider differencing or detrending.")
```

The Augmented Dickey-Fuller (ADF) test results indicate the following:

ADF Statistic: −10.25

p-value: 4.54×10−18(which is essentially zero for practical purposes).
    
**Interpretation**

Null Hypothesis: The data has a unit root (i.e., it is non-stationary).
    
Alternative Hypothesis: The data does not have a unit root (i.e., it is stationary).

Since the p-value is significantly less than 0.05, we reject the null hypothesis. This means data is stationary, and differencing or detrending is not required. We can Proceed directly with SARIMA or ARIMA modeling without additional transformations.

```python
# Assuming the 'Monthly Expenses' column is numeric after aggregation
df_final = df_final.dropna(subset=['Monthly Expenses'])  # Remove rows with missing values in 'Monthly Expenses'

# Fit ARIMA model
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(df_final['Monthly Expenses'], order=(1, 1, 1))  
model_fit = model.fit()

# Make forecasts
forecast = model_fit.forecast(steps=1)  # Forecast the next month
print(forecast)
```
